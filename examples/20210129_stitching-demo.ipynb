# Stitching of solar modules

This notebooks demonstrates the simple stitching that is provided with this toolbox.

```python
import pvinspect as pv
```

## Load data

This toolbox comes with a stitching demo dataset. This demo data is created by splitting images from `pvinspect.data.datasets.poly10x6` in two parts. So let's load five demo pairs:

```python
data = pv.data.datasets.stitching_demo(5)
```

So we now have the first pair of images as follows:

```python
data[0][0].show()
```

```python
data[0][1].show()
```

## Stitch

Now, stitching is as simple as :

```python
result = pv.preproc.stitching.stitch_images(data[0])
```

```python
result.show()
```

We can compare this to the original image as follows and find that the results are pretty similar:

```python
pv.data.datasets.poly10x6(1).head()
```
