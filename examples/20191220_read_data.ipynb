## Using `read_module_images`

[```read_module_images```](https://ma0ho.github.io/pvinspect/data/index.html#pvinspect.data.read_module_images) is used to read module images from the disk. You need to specify the data path, imaging modality and if the images originate from the same camera such that intrinsic parameters are the same for all images. In addition, you might also specify the number of cells in a row and in a column. For more information and additional arguments, please refer to the [documentation](https://ma0ho.github.io/pvinspect/data/index.html#pvinspect.data.read_module_images).

````python
imgs = pv.data.io.read_module_images('../pvinspect/data/datasets/20191219_poly10x6', pv.data.EL_IMAGE, same_camera=True, cols=10, rows=6)
````

We can now proceed with the resulting [```ModuleImageSequence```](https://ma0ho.github.io/pvinspect/data/image.html#pvinspect.data.image.ModuleImageSequence)..:

````python
imgs.head()
````

````python
len(imgs)
````

````python
imgs.tail()
````

Note that the **metadata** that we suppied during reading is stored with the images and can be accessed directly:

````python
imgs.meta_to_pandas()
````

## Using `read_images`

We can also use the more general purpose method `read_images` to load images with less meta data:

````python
imgs2 = pv.data.io.read_images('../pvinspect/data/datasets/20191219_poly10x6')
````

In this case, we have less meta data associated with the images an, in addition, we get an `ImageSequence` instead of a `ModuleImageSequence`:

````python
imgs2.meta_to_pandas()
````
